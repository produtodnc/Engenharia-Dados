```python
import pandas as pd

fact_reviews = pd.read_csv('../input-data/woman_clothing/Womens Clothing E-Commerce Reviews.csv', index_col=0)
fact_reviews.sample(5)
```

```python
fact_reviews.info()
```

```python
list(fact_reviews.columns)
```

```python
# Remover registros com valores nulos

fact_reviews = fact_reviews.dropna(
    axis=0,
    subset=['Clothing ID',
            'Title',
            'Review Text',
            'Rating']
)

fact_reviews
```

```python
fact_reviews.info()
```

```python
# normalizacao de texto dos reviews


def std_text(text):
    text = str(text)
    cleaned_text = ''.join(e for e in text if e.isalnum() or e.isspace())
    cleaned_text = cleaned_text.lower()

    return cleaned_text
```

```python
fact_reviews["Title"] = fact_reviews["Title"].apply(lambda x: std_text(x))
fact_reviews["Review Text"] = fact_reviews["Review Text"].apply(lambda x: std_text(x))
```

```python
fact_reviews
```

```python
# Carregando um segudo dataset com as categorias das rouptas

dim_categories = pd.read_csv('../input-data/woman_clothing/womens_departments.csv', index_col=0)
dim_categories
```

```python
dim_categories.info()
```

```python
# Removendo valores nulos

dim_categories = dim_categories.dropna(axis=0)
```

```python
dim_categories.info()
```

```python
# Fazendo o join entre as duas tabelas


merged_df = pd.merge(
    fact_reviews,
    dim_categories,
    on="Clothing ID",
    how="inner"
)
```

```python
merged_df.sample(10)
```

```python
merged_df.info()
```

```python
merged_df["Rating"].unique()
```

```python
merged_df = merged_df[merged_df["Rating"].isin(['3', '5', '2', '4', '1'])]
```

```python
merged_df["Rating"].unique()
```

```python
merged_df["Rating"] = merged_df["Rating"].astype(int)
```

```python
merged_df.info()
```

```python
# Agora podemos agrupar por categoria e calcular a media do Rating (Avaliacao)

avg_rating_by_category = merged_df.groupby("Class Name")["Rating"].mean()

avg_rating_by_category
```

```python
df_final_rating = pd.DataFrame(avg_rating_by_category).sort_values("Rating", ascending=False)
```

```python
df_final_rating.round(2)
```
