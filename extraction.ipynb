```python
# Agora precisamos listart os jogadores que jogam na La Liga (id = 140)

import requests

url = "https://api-football-v1.p.rapidapi.com/v3/players"

headers = {
	"X-RapidAPI-Key": "c36c4340d4msh804f051810841e8p188671jsnbf15af1b5bf5",
	"X-RapidAPI-Host": "api-football-v1.p.rapidapi.com"
}

raw_data = []

for page in range(1,20):
	querystring = {"league":"140","season":"2023", "page": page}
	response = requests.get(url, headers=headers, params=querystring)
	print(page, response.status_code)

	raw_data.append(response.json()["response"])
    
len(raw_data)
```

```python
raw_data[0]
```

```python
flatten_list = [x for xs in raw_data for x in xs]

len(flatten_list)
```

```python
flatten_list[0]
```

```python
players = []

for player in flatten_list:
    players.append(player["player"])
```
